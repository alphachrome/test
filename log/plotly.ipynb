```python
import time
import pandas as pd
import numpy as np 
import plotly.graph_objs as go
from plotly.offline import plot, iplot, init_notebook_mode
#from plotly.plotly import iplot, plot
from plotly import tools

app.css.append_css({"external_url": "https://codepen.io/chriddyp/pen/bWLwgP.css"})


init_notebook_mode()

def uj2w(t,e):
    w = [(e[1]-e[0])/(t[1]-t[0])/1e6]
    for t0,t1,e0,e1 in zip(t[:-1],t[1:],e[:-1],e[1:]):
        p = (e1-e0)/(t1-t0)/1e6
        if p>100 or p<-100:
            p=0
        w.append(p)
    return np.array(w)

PLOT_ALPHA=0.7

while True:
    df = pd.read_csv('/tmp/mon.csv')

    plt_w = uj2w(df['time'],df['cpu_uj'])
    sys_w = uj2w(df['time'],df['sys_uj'])*2

    #t = df['time']
    t = np.arange(len(df['time'])).astype(float)
    t = (t - t[0])/60/60

    pbat0 = df.i0*df.v0/1e12
    pbat1 = df.i1*df.v1/1e12

    trace_v0 = go.Scatter(x=t, y=df.v0/1e6, line = dict(width=1.5), name='BAT0', 
                    marker = dict(symbol='circle', color='orange'),
                    legendgroup='bat0', xaxis='TIME')
    trace_v1 = go.Scatter(x=t, y=df.v1/1e6, line = dict(width=1.5), name='BAT1',
                    marker = dict(symbol='circle', color='steelblue'), 
                    legendgroup='bat1')
    trace_i0 = go.Scatter(x=t, y=df.i0/1e6, line = dict(width=1.5), name='BAT0',
                    marker = dict(symbol='circle', color='orange'),
                    legendgroup='bat0', showlegend=False)
    trace_i1 = go.Scatter(x=t, y=df.i1/1e6, line = dict(width=1.5), name='BAT1',
                    marker = dict(symbol='circle', color='steelblue'), 
                    legendgroup='bat1', showlegend=False)
    trace_p0 = go.Scatter(x=t, y=pbat0, line = dict(width=1.5), name='BAT0',
                    marker = dict(symbol='circle', color='orange'),
                    legendgroup='bat0', showlegend=False)
    trace_p1 = go.Scatter(x=t, y=pbat1, line = dict(width=1.5), name='BAT1',
                    marker = dict(symbol='circle', color='steelblue'), 
                    legendgroup='bat1', showlegend=False)
    trace_sys = go.Scatter(x=t, y=sys_w, line = dict(width=1.5), name='PSYS', 
                    legendgroup='psys')
    trace_plt = go.Scatter(x=t, y=plt_w, line = dict(width=1.5), name='SOC',
                    legendgroup='plt')
    trace_rop = go.Scatter(x=t, y=sys_w-plt_w, line = dict(width=1.5), name='ROP',
                    legendgroup='rop')
    hist_p0 = go.Histogram(x=pbat0[pbat0>0], histnorm='probability', nbinsx=128, name='BAT0',
                    marker = dict(color='orange'), legendgroup='bat0', showlegend=False,
                    cumulative=dict(enabled=True))
    hist_p1 = go.Histogram(x=pbat1[pbat1>0], histnorm='probability', nbinsx=128, name='BAT1',
                    marker = dict(color='steelblue'), legendgroup='bat1', showlegend=False,
                    cumulative=dict(enabled=True))

    layout = go.Layout(

        width=1024, height=768, 
        margin=dict(l=40, b=40, r=0, t=40),
        xaxis1=dict(scaleanchor='x2'),
        xaxis3=dict(scaleanchor='x2'),
        xaxis4=dict(title='BAT0 POWER (W)'),
        xaxis5=dict(title='BAT1 POWER (W)'),
        yaxis1=dict(title='VOLTAGE (V)', range=([6,9])),
        yaxis2=dict(title='CURRENT (A)'),
        yaxis3=dict(title='POWER (W)'),
        yaxis4=dict(title='PROBABILITY')    
    )

    fig = tools.make_subplots(rows=4, cols=2, vertical_spacing=0.085,
        specs=[[{'colspan': 2},None], [{'colspan': 2},None], [{'colspan': 2},None], [{}, {}]],
        print_grid=True);

    fig.append_trace(trace_v0, 1, 1)
    fig.append_trace(trace_v1, 1, 1)
    fig.append_trace(trace_i0, 2, 1)
    fig.append_trace(trace_i1, 2, 1)
    fig.append_trace(trace_p0, 3, 1)
    fig.append_trace(trace_p1, 3, 1)
    fig.append_trace(trace_sys, 3, 1)
    fig.append_trace(trace_plt, 3, 1)
    fig.append_trace(trace_rop, 3, 1)
    fig.append_trace(hist_p0, 4, 1)
    fig.append_trace(hist_p1, 4, 2)

    fig.layout.update(layout)

    iplot(fig, filename='lux_2')
    
```
